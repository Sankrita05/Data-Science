## 5 number Summary
1. Minimum Value

3. Median
4. Q3- 75 percentile

```python
import numpy as np
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,1000,1100]
```

```python
## [Lower Fence<---> Higher Fence]
Q1=np.percentile(lst_marks,[25])
print(Q1)
```

```python
minimum,Q1,Q2,Q3,maximum=np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
```

```python
maximum
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_fence=Q1-1.5*(IQR)
higher_fence=Q3+1.5*(IQR)
```

```python
lower_fence,higher_fence
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Given data
marks = [22, 24, 26, 28, 29, 31, 35, 37, 41, 53, 64]

# Create box plot with filled outliers
plt.figure(figsize=(7, 4))
sns.boxplot(marks, orient='h', fill=False, 
            flierprops={"marker": "o", "markerfacecolor": "red", "markersize": 8})

# Labels and title
plt.xlabel("Marks")
plt.title("Box Plot with Filled Outliers")

# Show plot
plt.show()
```
